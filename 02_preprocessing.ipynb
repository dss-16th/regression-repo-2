### 3-4. LabelEncoder()
- 순서가 있는 변수: BusinessTravel, OverTime

```python
hr_raw = ibm_hr_raw.copy()
```

```python
hr_raw['BusinessTravel'].unique()
```

```python
hr_raw['OverTime'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(hr_raw['BusinessTravel'])
hr_raw['BusinessTravel'] = le.transform(hr_raw['BusinessTravel'])
le.fit(hr_raw['OverTime'])
hr_raw['OverTime'] = le.transform(hr_raw['OverTime'])

hr_raw.head()
```

```python
hr_raw[['BusinessTravel', 'OverTime']]
```

### 3-5. OneHotEncoder()
- 순서가 없는 변수: EducationField, Gender, JobRole, MaritalStatus

```python
hr_raw.EducationField.unique(), len(hr_raw.EducationField.unique())
```

```python
hr_raw.Gender.unique(), len(hr_raw.Gender.unique())
```

```python
hr_raw.JobRole.unique(), len(hr_raw.JobRole.unique())
```

```python
hr_raw.MaritalStatus.unique(), len(hr_raw.MaritalStatus.unique())
```

```python
# 원핫인코딩할 변수
col_names_encode = ['EducationField', 'Gender', 'JobRole', 'MaritalStatus']
```

```python
df_encode = hr_raw[col_names_encode]
df_encode
```

```python
# 새로 생길 컬럼은 20개
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
enc.fit(df_encode)
df_encode = pd.DataFrame(data=enc.transform(df_encode), columns=enc.get_feature_names(col_names_encode), dtype=bool)
# trasfer true and false to 1 and 0
df_encode = df_encode * 1
```

```python
df_encode
```

```python
hr_raw
```

```python
# 데이터 합치기
hr_raw = pd.concat((hr_raw, df_encode), axis=1)
hr_raw = hr_raw.drop(columns=col_names_encode)
hr_raw
```

```python
len(hr_raw.columns)
```

```python
hr_raw.columns
```

```python
hr_raw.info()
```

```python
# 컬럼명 정리
```

```python
hr_raw.columns = hr_raw.columns.str.replace(' ', '')
```

```python
hr_raw.columns = hr_raw.columns.str.replace('&', '')
```

```python
hr_raw.columns
```

```python
print("Before : \n", list(ibm_hr_raw.columns), "\n", f"총 {len(ibm_hr_raw.columns)}개", "\n")
print("After : \n", list(hr_raw.columns), "\n", f"총 {len(hr_raw.columns)}개")
```

```python
# 인코딩 버전 저장
hr_raw.to_csv('datas/hr_raw.csv')
```
